```python
import pandas as pd
import os
```

Reading files in the directory By OS

```python
files = [file for file in os.listdir('/home/ducati/Documents/python/Pandas-Data-Science-Tasks/SalesAnalysis/Sales_Data')]
all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv('/home/ducati/Documents/python/Pandas-Data-Science-Tasks/SalesAnalysis/Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("alldata.csv", index=False)
```
